```python
%matplotlib inline
import numpy as np 
import pandas as pd
import seaborn as sns
sns.set(style='darkgrid')
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Loan_data.csv")
```

```python
df.head()
```

```python
df=df.dropna()
```

```python
df.reset_index(drop=True)
```

```python
df.dtypes
```

```python
df['Dependents'] = df['Dependents'].str.rstrip('+')
df['Dependents'] = df['Dependents'].astype(np.int32)
df['Gender'] = df['Gender'].map({'Female':0,'Male':1}).astype(np.int32)
df['Married'] = df['Married'].map({'No':0, 'Yes':1}).astype(np.int32)
df['Education'] = df['Education'].map({'Not Graduate':0, 'Graduate':1}).astype(np.int32)
df['Self_Employed'] = df['Self_Employed'].map({'No':0, 'Yes':1}).astype(np.int32)
df['Loan_Status'] = df['Loan_Status'].map({'N':0, 'Y':1}).astype(np.int32)
df['Property_Area'] = df['Property_Area'].map({'Rural':0, 'Urban':1,'Semiurban':2}).astype(np.int32)
```

```python
df.head(100)
```

```python
df.dtypes
```

Dividing the given train dataset in the form of features and target variables 

```python
X,y  = df.iloc[:, 1:-1], df.iloc[:, -1]
```

```python
df.corr()
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
```

```python
# We split the X into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
#C is inverse to lambda .
#It is the hyperparameter which controls the regularization basically to avoid overfitting the model.
from sklearn.linear_model import LogisticRegression
LR_model = LogisticRegression(C=0.01).fit(X_train,y_train)
LR_model
```

```python
y_test=y_test.ravel()
y_test
```

```python
y_pred = LR_model.predict(X_test)
y_pred=y_pred.ravel()
y_pred
```

```python
#Predict will give either 0 or 1 as output
#Predict_proba will give the only probability of occurence of 0 and 1.
LR_yhat_prob = LR_model.predict_proba(X_test)
LR_yhat_prob
```

```python
LR_yhat_prob.shape
```

```python
#The F-score, also called the F1-score, is a measure of a model’s accuracy on a dataset
#The F-score is a way of combining the precision and recall of the model, and it is defined as the harmonic mean of the model’s precision and recall.
from sklearn.metrics import f1_score
print("Logistic Regression F1-score: %.2f" % f1_score(y_test, y_pred) )
```

```python
from sklearn.metrics import mean_absolute_error
#evaluation predictions
mae=mean_absolute_error(y_test,y_pred)
print('MAE:%.3f'% mae)
```

```python
#Log-loss is indicative of how close the prediction probability is to the corresponding actual/true value (0 or 1 in case of binary classification). 
#The more the predicted probability diverges from the actual value, the higher is the log-loss value.
from sklearn.metrics import log_loss
"LR LogLoss:%.2f"%(log_loss(y_pred, LR_yhat_prob))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
confusion_matrix
#The result is telling us that we have 12+59=71 correct predictions and 2+23=25 wrong predictions.(71+25=96)
```
